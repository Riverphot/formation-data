```python
#Fonction d'analyse exploratoire des fichiers en lignes
def analyze_columns_line_by_line(df):    
    print("\nüìä **Analyse Exploratoire du DataFrame** üìä\n")
    for col in df.columns:
        print(f"üîπ **Colonne :** {col}")
        print(f"   - Type de donn√©es : {df[col].dtype}")
        print(f"   - Valeurs manquantes : {df[col].isnull().sum()}")
        print(f"   - Unique : {df[col].nunique(dropna=True)}")
        print(f"   - Distinct : {(df[col].value_counts() == 1).sum()}")
        print(f"   - Min : {df[col].dropna().min()}")
        print(f"   - Max : {df[col].dropna().max()}")
        print("-" * 50)  # S√©parateur visuel entre les colonnes
```

<div style="background:#FFEEDA;padding:10px;">
Les donn√©es sont pr√©sentes pour plusieurs ann√©es. Nous souhaitons connaitre les donn√©es les plus r√©centes par pays pour connaitre la population actuelle. Pour cela nous allons d√©terminer l'ann√©e la plus r√©cente ou l'information est fournie pour un pays.
    <br/>Dans un second temps nous voulons d√©terminer la croissance d√©mographique de la population de chaque pays.
</div>

```python
#On v√©rifie les ann√©es disponibles par pays
df_population_pivot = df_population.pivot(index="Zone", columns="Ann√©e", values="Population") \
                                   .rename_axis(None, axis=1) \
                                   .reset_index()
df_population_pivot.head()
```

```python
# On v√©rifie si il y a des valeurs manquantes
df_population_nan = df_population_pivot[df_population_pivot.isna().any(axis=1)]
df_population_nan.head(20)
```

<div style="background:#FFEEDA;padding:10px;">
Les donn√©es maquantes peuvent √™tres corrig√©es ou peuvent √™tre ignor√©e dans le calcul de la croissance de la population.
On s√©lectionnera alors l'ann√©e la plus ancienne et la plus r√©cente disponible pour chaque pays.
</div>

```python
#On compl√®te les informations pour la R√©publique de Cor√©e (avec des sources nouvelles)
#Dans notre analyse on pourrait aussi choisir d'exclure ce pays (sa sitiation politique l'exclut de base comme candidat)
df_population_pivot.loc["R√©publique populaire d√©mocratique de Cor√©e", 2019] = 26040000
df_population_pivot.loc["R√©publique populaire d√©mocratique de Cor√©e", 2020] = 26140000
df_population_pivot.loc["R√©publique populaire d√©mocratique de Cor√©e", 2021] = 26230000
df_population_pivot.loc["R√©publique populaire d√©mocratique de Cor√©e", 2022] = 26330000
```

```python
# S√©lection des colonnes contenant les ann√©es
annees_colonnes = [col for col in df_population_pivot.columns if isinstance(col, int)]

# D√©terminer l'ann√©e la plus ancienne et la plus r√©cente pour chaque pays
df_population_pivot["Ann√©e_min"] = df_population_pivot[annees_colonnes].apply(lambda row: min(row.dropna().index), axis=1)
df_population_pivot["Ann√©e_max"] = df_population_pivot[annees_colonnes].apply(lambda row: max(row.dropna().index), axis=1)

# Extraire la population initiale et finale
df_population_pivot["Pop_initiale"] = df_population_pivot.apply(lambda row: row[row["Ann√©e_min"]], axis=1)
df_population_pivot["Population"] = df_population_pivot.apply(lambda row: row[row["Ann√©e_max"]], axis=1)

# Calcul du nombre d'ann√©es d'intervalle
df_population_pivot["N"] = df_population_pivot["Ann√©e_max"] - df_population_pivot["Ann√©e_min"]

# Calcul de la croissance moyenne annuelle (%) sur l'intervalle disponible
df_population_pivot["Croissance population (%)"] = (((df_population_pivot["Population"] / df_population_pivot["Pop_initiale"]) ** (1 / df_population_pivot["N"]) - 1) * 100).round(2)

# Affichage du tableau final avec la croissance pour tous les pays
df_population_pivot.head()
```

```python
#On conserve uniquement les colonnes pertinentes
df_population_pivot = df_population_pivot.drop(columns=[2018, 2019, 2020, 2021, 2022, "Ann√©e_min", "Ann√©e_max", "Pop_initiale", "N"])
```

```python
# Affichage des premi√®res lignes 
df_population_pivot.head()
```

```python
#On compl√®te les valeurs manquantes si disponible

df_analyse.loc[df_analyse["Zone"] == "Chine", "Stabilit√© Politique"] = -0.26

df_analyse.loc[df_analyse["Zone"] == "Chine, Taiwan Province de", "Stabilit√© Politique"] = 0.85
df_analyse.loc[df_analyse["Zone"] == "Chine, Taiwan Province de", "Pib/habitant"] = 1

df_analyse.loc[df_analyse["Zone"] == "Chine, continentale", "Stabilit√© Politique"] = -0.26

#Il s'agit de t√©rritoires fran√ßais (attention on ne trouvera que des r√©sultats fran√ßais mais ces r√©gions peuvent avoir leur particularit√©)
df_analyse.loc[df_analyse["Zone"] == "Nouvelle-Cal√©donie", "Stabilit√© Politique"] = 0.33
df_analyse.loc[df_analyse["Zone"] == "Nouvelle-Cal√©donie", "Pib/habitant"] = 54988.3
df_analyse.loc[df_analyse["Zone"] == "Polyn√©sie fran√ßaise", "Stabilit√© Politique"] = 0.33
df_analyse.loc[df_analyse["Zone"] == "Polyn√©sie fran√ßaise", "Pib/habitant"] = 54988.3
```

<div style="background-color:#638BC9;" >
<h2 style="margin: auto; padding: 20px; color:#fff; ">√âtape 4 - Cr√©ation de variables compl√©mentaires</h2>
</div>

<div style="background:#FFEEDA;padding:10px;">
Pour compl√©ter les informations de nos dataframes de nouvelles variables (indicateurs) vont √™tre cr√©es √† partir des donn√©es pr√©sentes : 
    <ul>
        <li><strong>Croissance d√©mographique : </strong>Indicateur de la croissance moyenne annuelle du pays. Indicateur de l'√©volution √† venir du march√©.</li>
        <li><strong>Taux de stockage : </strong>Mesure la propension d‚Äôun pays √† stocker ou consommer rapidement sa production et ses importations.</li>
        <li><strong>Taux d‚Äôautosuffisance (%) : </strong>Indique si un pays produit suffisamment pour couvrir sa consommation. </li>
        <li><strong>Taux de d√©pendance aux importations (%) : </strong>Mesure la d√©pendance d‚Äôun pays aux importations pour sa consommation de volaille. </li>
    </ul>
</div>

<div style="background:#C0DFF4;" >
<h3 style="margin: auto; padding: 20px; color: #0D2E63; ">4.1 - Cr√©ation de la variable Croissance d√©mographique</h3>
</div>

<div style="background:#FFEEDA;padding:10px;">
Cette variable est une mesure bas√© sur une moyenne exponentielle de l'√©volution de la d√©mographie au cours des derni√®res ann√©es. Elle permet de se faire une id√©e de l'√©volution en % de la population pour une ann√©e √† venir. Elle est un bon indicateur de la croissance d'un march√© pour un pays.
    <br/><em>Cette valeur a √©t√© calcul√©e pr√©c√©demmet dans le df_population : Croissance annuelle population (%).</em>
</div>

<div style="background:#C0DFF4;" >
<h3 style="margin: auto; padding: 20px; color: #0D2E63; ">4.2 - Cr√©ation de la variable Taux de stockage</h3>
</div>

<div style="background:#FFEEDA;padding:10px;">
Le Taux de stockage mesure quelle proportion de la disponibilit√© int√©rieure est conserv√©e sous forme de stock plut√¥t que consomm√©e ou export√©e imm√©diatement.
    <br/>Taux de stockage = Variation de stock / Disponibilit√© Int√©rieure
    <br/><em>Ce pourcentage repr√©sente la part de l‚Äôoffre alimentaire qui est mise en r√©serve au lieu d‚Äô√™tre directement utilis√©e.</em>
</div>

```python
# Calcul du Taux de stockage
df_analyse["Taux stockage (%)"] = (df_analyse["Variation stock (1000t)"] / df_analyse["Disponibilit√© int√©rieure (1000t)"] * 100).round(2)

# Affichage des premi√®res lignes du DataFrame
df_analyse.head()
```

```python
#On v√©rifie si il y a des informations manquantes dans le dataframe global
df_analyse_incomplet = df_analyse[df_analyse.isna().any(axis=1)]

# Afficher les premi√®res lignes du DataFrame r√©sultant
df_analyse_incomplet.head(50)
```

Les informations sur ce pays semblent peut fiable, on a bien une dispo par kg/personne/an mais avec des production a 0, imort 0, bref des donn√©es incompl√®tes. On supprime donc ce pays.
    <br/><em>Suppression sans regret de la liste, Tuvalu est une petite archipel ind√©pendante en Oc√©anie, proche de l'extension en tant qu'√©tat ind√©pendant.</em>

<div style="background:#C0DFF4;" >
<h3 style="margin: auto; padding: 20px; color: #0D2E63; ">4.3 - Cr√©ation de la variable Taux d‚Äôautosuffisance </h3>
</div>

<div style="background:#FFEEDA;padding:10px;">
Ce taux mesure quelle proportion de la disponibilit√© int√©rieure est couverte par la production locale.
    <br/>Taux¬†d‚Äôautosuffisance= Production¬†/ Disponibilit√© int√©rieur
    <br/><em>Ce pourcentage repr√©sente la part de la consommation int√©rieure qui est couverte par la production nationale.</em>
</div>

```python
# Calcul du Taux d‚Äôautosuffisance en pourcentage
df_analyse["Taux autosuffisance (%)"] = ((df_analyse["Production (1000t)"] / df_analyse["Disponibilit√© int√©rieure (1000t)"]) * 100).round(2)

# Afficher dataframe
df_analyse.head()
```

```python
#On v√©rifie si il y a des informations manquantes dans le dataframe global
df_analyse_incomplet = df_analyse[df_analyse.isna().any(axis=1)]

# Afficher les premi√®res lignes du DataFrame r√©sultant
df_analyse_incomplet.head(50)
```

<div style="background:#C0DFF4;" >
<h3 style="margin: auto; padding: 20px; color: #0D2E63; ">4.4 - Cr√©ation de la variable Taux de d√©pendance aux importations </h3>
</div>

<div style="background:#FFEEDA;padding:10px;">
Ce taux repr√©sente la part de la consommation int√©rieure qui est couverte par les importations plut√¥t que par la production nationale.
    <br/>Taux¬†de¬†d√©pendance¬†aux¬†importations = Importations¬†/ Disponibilit√© int√©rieur
    <br/><em>Ce pourcentage repr√©sente la part de la consommation int√©rieure qui est couverte par les importations plut√¥t que par la production nationale.</em>
</div>

```python
# Calcul du Taux de d√©pendance aux importations
df_analyse["Taux d√©pendance importations (%)"] = ((df_analyse["Importations (1000t)"] / df_analyse["Disponibilit√© int√©rieure (1000t)"]) * 100).round(2)

# Afficher dataframe
df_analyse.head()
```

```python
#On v√©rifie si il y a des informations manquantes dans le dataframe global
df_analyse_incomplet = df_analyse[df_analyse.isna().any(axis=1)]

# Afficher les premi√®res lignes du DataFrame r√©sultant
df_analyse_incomplet.head(50)
```
